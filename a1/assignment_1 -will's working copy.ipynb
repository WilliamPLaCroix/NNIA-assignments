```python
# Numpy
array = np.random.randint(21, 25, 30)
print(array)
mean = np.mean(array)
std = np.std(array)
print(f"Mean: {mean}, Std: {std}")

def mean(array: list) -> float:
    return sum(array) / len(array)

def std(array: list) -> float:
    return np.sqrt(sum((array - mean(array)) ** 2) / len(array))

print(f"Mean: {mean(array)}, Std: {std(array)}")
```

## 3.2.1
For this exercise you will be working with tokenized phrases extracted from the ’sms_spam’ database from
the ’Huggingface’ platform (dataset 1). Your task is to:
* complete the skeleton code for the pytorch dataset subclass
* create a dataloader instance with a batch size of 1
* iterate over the dataloader and print the labels and features of each batch

```python
from torch.utils.data import Dataset, DataLoader

class MyDataset(Dataset):
    def __init__(self, path):
        # Read the json file and store the data
        self.data = pd.read_json(path)

    def __len__(self):
        # Return the number of rows as the size of the dataset
        self.size = len(self.data)
        return self.size

    def __getitem__(self, index):
        # Return the features and label of a datapoint using its index
        features = self.data.iloc[index]["tokens"]
        label = self.data.iloc[index]["label"]

        return features, label
```

```python
# Create an instance of the dataset class with the path to the json file
dataset = MyDataset("data/Dataset1/data.json")

# Create a dataloader instance and set the batch size to 1 and shuffle to true
dataloader = DataLoader(dataset, batch_size=1, shuffle=True)

for features, labels in dataloader:
    print(features, labels)
```
