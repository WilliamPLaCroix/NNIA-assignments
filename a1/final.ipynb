### Numpy

```python
import math
array = np.random.randint(21, 25, 30)
print(f"The array: {array}")
mean = np.mean(array)
std = np.std(array)
print(f"Mean: {mean}; Std: {std}")

def my_mean(array: object) -> float:
    """ Custom implementation of the numpy.mean function. Sums up all elements of the array and divides by the count of elements.
    :param:     array: a numpy array of numbers (in this case integers).
    :return:    the mean of the array, in most cases a float.
    """
    return sum(array) / len(array)

def my_std(array: object) -> float:
    """ Custom implementation of the numpy.std function. It calculates the sum of the squared errors, divides by the count of the elements and takes the squareroot.
    :param:     array: a numpy array of numbers (in this case integers).
    :return:    the standard deviation of the array, in most cases a float.
    """
    return math.sqrt(sum((array - my_mean(array)) ** 2) / len(array))

custom_mean = my_mean(array)
custom_std = my_std(array)
print(f"Mean: {custom_mean}; Std: {custom_std}")
print(f"Are they the same? Answer: {(mean == custom_mean) and (std == custom_std)}")
```

##### Observation:
Sometimes the std value is a tiny bit off at the smallest digit(s), due to rounding errors.

## 3.2.1
For this exercise you will be working with tokenized phrases extracted from the ’sms_spam’ database from
the ’Huggingface’ platform (dataset 1). Your task is to:
* complete the skeleton code for the pytorch dataset subclass
* create a dataloader instance with a batch size of 1
* iterate over the dataloader and print the labels and features of each batch

```python
from torch.utils.data import Dataset, DataLoader

class MyDataset(Dataset):
    def __init__(self, path: str) -> None:
        """ Constructor of the MyDataset class. Reads the json file and stores the data.
        :param:     path: string, filename of the dataset stored as a json file.
        :return:    None (Constructor)
        """
        self.data = pd.read_json(path)

    def __len__(self) -> int:
        """ Implements the len-operator. Returns the number of rows as the size of the dataset
        :param:     None
        :return:    len(self.data): int, number of rows.
        """
        return len(self.data)

    def __getitem__(self, index: int) -> (object, object):
        """ Implements the bracket-operator[]. Returns the features and label of a datapoint using its index.
        :param:     index: int, index of element.
        :return:    (features: numpy.int64, label: list): taken from the original dataset.
        """
        if (index < 0) or (index >= len(self)):
            raise IndexError("Index out of range!")
        features = self.data.iloc[index]["tokens"]
        label = self.data.iloc[index]["label"]
        return features, label
```

```python
# Create an instance of the dataset class with the path to the json file
dataset = MyDataset("data/Dataset1/data.json")

# Create a dataloader instance and set the batch size to 1 and shuffle to true
dataloader = DataLoader(dataset, batch_size=1, shuffle=True)

# Iterate over the dataloader and print the features and labels of each batch
for i in range(len(dataloader)):
    print("###############################")
    print(f"Batch No.: {i+1} / {len(dataloader)}")
    print(f"Features: {dataloader.dataset[i][0]}")
    print(f"Label: {dataloader.dataset[i][1]}")

    
    
    
#####################################
# Ugly output, but shorter version: #
#####################################

#for batch in dataloader:
#    print(batch)
```

# 3.2.2 (3 points)
For the second exercise you will be working with a csv file containing information on the symptoms and prognosis related to 11 vector-borne
diseases. The symptoms are the features and the prognosis serves as the label. For this task you will also need to perform [one-hot encoding](https://machinelearningmastery.com/why-one-hot-encode-data-in-machine-learning/) on the prognosis column to transform it into numeric values. You can create your own function for this or use or use a function from an external library such as [sci-kit's LabelBinerizer](https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.LabelBinarizer.html). Your task is again to:
* create a pytorch dataset subclass
* create a dataloader instance with a batch size of 32
* iterate over the dataloader and print the labels and features of each batch

```python
from sklearn.preprocessing import LabelBinarizer

class MyDataset_csv(LabelBinarizer):
    def __init__(self, path: str) -> None:
        """ Constructor of the MyDataset_csv class. Reads the json file and stores the data. The LabelBinarizer determines the set of distinct classes and transforms each label into a onehot-vector.
        :param:     path: string, filename of the dataset stored as a csv file.
        :return:    None (Constructor)
        """
        self.data = pd.read_csv(path)
        self.features = self.data.drop('prognosis', axis=1)
        self.labels = self.data[['prognosis']]
        self.lb = LabelBinarizer()
        self.lb.fit(self.labels)
        self.labels = self.lb.transform(self.labels)
        
    def __len__(self) -> int:
        """ Implements the len-operator. Returns the number of rows as the size of the dataset.
        :param:     None
        :return:    len(self.data): int, number of rows.
        """
        return len(self.data)
    
    def __getitem__(self, index: int) -> (object, object):
        """ Implements the bracket-operator[]. Returns the features and label of a datapoint using its index.
        :param:     index: int, index of element.
        :return:    (features, labels): a tuple of tensors taken from the original dataset.
        """
        if (index < 0) or (index >= len(self)):
            raise IndexError("Index out of range!")
        features = self.features.iloc[index].to_numpy()
        labels = self.labels[index]
        return features, labels
```

```python
dataset_csv = MyDataset_csv("data/Dataset2/data.csv")

dataloader_csv = DataLoader(dataset_csv, batch_size=32)

# Iterate over the dataloader and print the features and labels of each batch
for i, batch in enumerate(dataloader_csv):
    print("###############################")
    print(f"Batch No.: {i+1} / {len(dataloader_csv)}")
    print(batch)
```
