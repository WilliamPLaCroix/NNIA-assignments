## 1.2 Multiplication properties and types of matrices (4 point)
Provide a definition and example for the following matrices:
1. Transpose of a matrix: effectively swaps the indices of a matrix $\vect{A}_{ij}\to \vect{A}_{ji}$:
$$\vect{A}=\begin{bmatrix} 1 & 2 & 3 \\ 4 & 5 & 6 \end{bmatrix} \to \vect{A}^T=\begin{bmatrix} 1 & 4 \\ 2 & 5 \\ 3 & 6 \end{bmatrix}$$

2. Identity matrix: a square matrix with 1s on the diagonal and 0s everywhere else:
$$\vect{I}=\begin{bmatrix} 1 & 0 & 0 \\ 0 & 1 & 0 \\ 0 & 0 & 1\end{bmatrix}$$

3. Inverse of a matrix: $\vect{A}$ inverse $(\vect{A}^{-1})$ is defined such that $\vect{A}\cdot\vect{A}^{-1}=\vect{I}$:
$$\begin{bmatrix} 4 & 3 \\ 3 & 2 \end{bmatrix} \cdot \begin{bmatrix} -2 & 3 \\ 3 & -4 \end{bmatrix}=\begin{bmatrix} 1 & 0 \\ 0 & 1 \end{bmatrix}$$

4. Symmetric matrix: a matrix is symmetric if it is equal to its transpose (mirrored around the diagonal):
$$\vect{A}=\begin{bmatrix} 1 & 2 & 3 \\ 2 & 4 & 6 \\ 3 & 6 & 12\end{bmatrix}=\vect{A}^T$$


5. Orthogonal matrix: a square matrix where its inverse is equal to its transpose $(\vect{A}^{-1}=\vect{A}^T)$
$$\vect{A}=\begin{bmatrix} -1 & 0 \\ 0 & 1\end{bmatrix}=\vect{A}^T=\vect{A}^{-1};$$
$$\vect{A}\cdot\vect{A}^{-1}=\begin{bmatrix} 1 & 0 \\ 0 & 1\end{bmatrix}$$

### 2.1 Verify 1.1 with numpy and PyTorch

```python
import numpy as np

def numpy_verify(A, B):
    try:
        print("A dot B =\n", np.matmul(A, B))
        print("")
    except ValueError:
        print("A dot B is not defined for")
        print("A =\n", A)
        print("and")
        print("B =\n", B)
        print("")

# 1.1.1
A = np.array([[1, 1], [2, 2]])
B = np.array([[2, -1], [2, -1]])
numpy_verify(A, B)

# 1.1.2
A = np.array([[2, -1], [2, -1]])
B = np.array([[1, 1], [2, 2]])
numpy_verify(A, B)

# 1.1.3
A = np.array([[1, 2, -1], [1, 0, 1]])
B = np.array([[1, 3, -2], [2, -2, 0], [4, -1, 2]])
numpy_verify(A, B)

# 1.1.4
A = np.array([[-1, -3, 5], [3, 4, 0]])
B = np.array([[4, 6], [0, 5]])
numpy_verify(A, B)
```

```python
import torch

def torch_verify(A, B):
    A = torch.Tensor(A)
    B = torch.Tensor(B)

    try:
        print("A dot B =\n", torch.mm(A, B))
        print("")
    except RuntimeError:
        print("A dot B is not defined for")
        print("A =\n", A)
        print("and")
        print("B =\n", B)
        print("")

# 1.1.1
A = np.array([[1, 1], [2, 2]])
B = np.array([[2, -1], [2, -1]])
torch_verify(A, B)

# 1.1.2
A = np.array([[2, -1], [2, -1]])
B = np.array([[1, 1], [2, 2]])
torch_verify(A, B)

# 1.1.3
A = np.array([[1, 2, -1], [1, 0, 1]])
B = np.array([[1, 3, -2], [2, -2, 0], [4, -1, 2]])
torch_verify(A, B)

# 1.1.4
A = np.array([[-1, -3, 5], [3, 4, 0]])
B = np.array([[4, 6], [0, 5]])
torch_verify(A, B)
```
